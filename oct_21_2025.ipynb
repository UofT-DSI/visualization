```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, size=50)

font1 = {'family': 'serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'serif', 'color': 'green', 'size': 14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x, y)

ax.set_title("Random Integer Plot", fontdict=font1)
ax.set_xlabel("X-axis Label", fontdict=font2)
ax.set_ylabel("Y-axis Label", fontdict=font2)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y, marker = '*', color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, 
marker = '*', 
markersize=12,
color = 'indigo', 
linestyle='--', 
linewidth=2,
markeredgecolor="#cf59fa",  # outline color
markerfacecolor="#FAFAF8")  # fill color
```

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
df = sns.load_dataset('iris')
```

```python
# Usual boxplot
ax = sns.boxplot(x='species', y='sepal_length', data=df)

# Add jitter with the swarmplot function
ax = sns.swarmplot(x='species', y='sepal_length', data=df, color="darkblue")
plt.show()
```
