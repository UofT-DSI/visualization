## accessible vs non-accessible viz

```python
# load libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set data and variables
# modified from: https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_colors.html

x1 = ['apple', 'blueberry', 'cherry', 'orange']
y1 = [40, 100, 30, 55]
hatches = ['+', '/', '||', '*']
bar_colors = ['red', 'blue', 'maroon', 'orange']
bar_labels = ['apple', 'blueberry', 'cherry', 'orange']

x2 = ['apple', 'blueberry', 'cherry', 'orange']
y2 = [40, 100, 30, 55]
bar2_colors = ['red', 'blue', 'maroon', 'orange']
bar2_labels = ['apple', 'blueberry', 'cherry', 'orange']

# plot figure

fig, (ax1,ax2) = plt.subplots(figsize=(10,4),ncols=2,nrows = 1)

ax1.bar(x1, y1, color=bar_colors, hatch=hatches, label=bar_labels)
ax2.bar(x2, y2, label=bar2_labels, color=bar2_colors)

ax1.set_ylabel('fruit supply')
ax1.set_title('Fruit supply by kind and color')
ax1.legend()


ax2.set_ylabel('fruit supply')
ax2.set_title('Fruit supply by kind and color')
ax2.legend()

## note that ax1 uses patterns while ax2 only relies on colour

```

```python
# sample data
np.random.seed(613)
x = np.arange(5)
y1 = np.random.randint(0,20,5)
y2 = np.random.randint(0,20,5)
y3 = np.random.randint(0,20,5)

# set colorblind friendly palette
plt.style.use('tableau-colorblind10')

# create line plot with 3 lines
fig, (ax1,ax2) = plt.subplots(ncols = 2, nrows = 1, figsize = (10,4))


ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

ax2.plot(x,y1, label = 'Person 1', marker = '>',
        markersize = 10)
ax2.plot(x,y2, label = 'Person 2', marker = '*',
        markersize = 10)
ax2.plot(x,y3, label = 'Person 3', marker = '+',
        markersize = 10)

# annotate y3 points
for i, txt in enumerate(zip(x, y3)):
    plt.annotate(txt, (x[i], y3[i]+0.5), ha='center')

# add legend
ax2.legend(loc='lower right')

# note that ax1 does not have labels for each point, ax2 has labels for y3 data
# note that both axes use a 'colorblind' palette


```

```python
# sample data
np.random.seed(613)
x = np.arange(5)
y1 = np.random.randint(0,20,5)
y2 = np.random.randint(0,20,5)
y3 = np.random.randint(0,20,5)

# set colorblind friendly palette
plt.style.use('default')

# create line plot with 3 lines
fig, (ax1,ax2) = plt.subplots(ncols = 2, nrows = 1, figsize = (10,4))


ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

ax2.plot(x,y1, label = 'Person 1', marker = '>',
        markersize = 10)
ax2.plot(x,y2, label = 'Person 2', marker = '*',
        markersize = 10)
ax2.plot(x,y3, label = 'Person 3', marker = '+',
        markersize = 10)

# annotate y3 points
for i, txt in enumerate(zip(x, y3)):
    plt.annotate(txt, (x[i], y3[i]+0.5), ha='center')

# add legend
ax2.legend(loc='lower right')


# note that ax1 does not have labels for each point, ax2 has labels for y3 data
# note that both axes do not use a 'colorblind' palette but the default

```
