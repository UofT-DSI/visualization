```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Setting a seed for reproducibility
np.random.seed(613)

# Generating data for reproducibility
x = np.arange(50)
y = np.random.randint(0,100,50)

# Creating a figure and an axes
fig, ax = plt.subplots (figsize = (5,3))

# Plotting a scatter plot on the axes
ax.scatter(x,y)
```

```python
# A bar chart on the axes
fig, ax = plt.subplots (figsize = (5,3))
ax.bar(x,y)
```

```python
# A line plot on the axes
fig, ax = plt.subplots (figsize = (5,3))
ax.plot(x,y)
```

```python
# A histogram on the axes
fig, ax = plt.subplots (figsize = (5,3))
ax.hist(y)
```

```python

# Line plot on the axes
fig, ax = plt.subplots(figsize=(5, 3))

# Create a histogram using y data
ax.plot(x,y)

# Add titles and labels
ax.set_title('Total growth over time')
ax.set_xlabel('Years since start')
ax.set_ylabel('Total growth')
fig.tight_layout
```

```python
# Defining custom for a font properties
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

# Creating a figure and an axes
fig, ax = plt.subplots(figsize=(5, 3))

# Creating a plot using x and y data
ax.plot(x, y)

# Adding titles and labels
ax.set_title('Total growth over time', fontdict=font1, loc = 'right') #Add moving title right
ax.set_xlabel('Years since start', fontdict=font2)
ax.set_ylabel('Total growth', fontdict=font2, loc = 'top') #Add moving title top

# Show the plot
plt.show()
```

```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt

# Generating data for the lollipop plot
x = np.arange(1, 21)
y = np.random.randint(1, 100, 20)

# Creating a figure and an axes
fig, ax = plt.subplots(figsize=(10, 6))

# Creating the lollipop plot
ax.stem(x, y, linefmt='--', markerfmt='o', basefmt=' ')

# Adding titles and labels
ax.set_title('Lollipop Plot', fontsize=16, color='blue')
ax.set_xlabel('X-axis', fontsize=12, color='green')
ax.set_ylabel('Y-axis', fontsize=12, color='green')

# Customizing the marker
markers, stemlines, baseline = ax.stem(x, y)
plt.setp(markers, marker='o', markersize=8, markeredgewidth=2, color='#fa9359', markeredgecolor='#7425b9')

# Display the plot
plt.show()
```
