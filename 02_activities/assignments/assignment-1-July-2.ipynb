```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613)  # For reproducibility
x = np.arange(50)
y = np.random.randint(0, 100, size=50)  


fig, ax = plt.subplots(figsize=(5, 3))
#ax.scatter(x,y)
#ax.bar(x,y)
#ax.barh(x,y)
#ax.plot(y)
ax.hist(y)
```

```python
font1 = {'family': 'serif',
         'color':  'blue',
         'weight': 'normal',
         'size': 16,
}
font2 = {'family': 'monospace',
         'color':  'green',
         'size': 14,
}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)

ax.set_title('total growth over time',fontdict=font1, loc = 'left', pad=20)
ax.set_ylabel('total growth',fontdict=font2)             
ax.set_xlabel('years since start',fontdict=font2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,
           y,
           marker='*',
           color='indigo')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(
	x, y,
	marker='.',
	color="#7425B0",
	linestyle='-',
	linewidth=2,
	markersize=12,
	markeredgecolor='#000000',
	markerfacecolor='#FA9359'
)
ax.grid(axis = 'y')
```

```python
import seaborn as sns

df = sns.load_dataset('iris')
sns.set_style("darkgrid")
sns.kdeplot(x=df.sepal_width, y=df.sepal_length)
plt.show()
sns.set_style("white")
sns.kdeplot(
    x=df.sepal_width, 
    y=df.sepal_length, 
    cmap="rainbow", 
    fill=True, 
    alpha=0.8, 
    bw_adjust=0.2, 
    thresh=0.01, 
    levels=20, 
    linewidths=3, 
    linestyles="dashed"
)
plt.scatter(df.sepal_width, df.sepal_length, c=np.random.rand(len(df)), s=200, marker='X', edgecolor='black', alpha=0.7)
plt.title("X PLOT!", fontsize=24, color='magenta', fontweight='bold')
plt.xlabel("Sepal Width", fontsize=16, color='red')
plt.ylabel("Sepal Length", fontsize=16, color='blue')
plt.show()
sns.set_style("white")
sns.kdeplot(x=df.sepal_width, y=df.sepal_length,
            cmap="Blues", fill=True, bw_adjust=0.5)
plt.show()
```
