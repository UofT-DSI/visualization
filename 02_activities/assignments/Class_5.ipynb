```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
##Basic Plot

sns.lineplot(data=tips, # choose our dataset
x='total_bill'
, # define our x variable
y='tip') # define our y variable
```

```python
sns.set_style('whitegrid')
```

```python
tipgraph = sns.lineplot(data=tips,
x='total_bill'
,
y='tip')
tipgraph.set(title='Tips vs. Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)')
```

```python
fig = plt.subplots(figsize=(10, 3))

tipgraph = sns.lineplot(data=tips,
    x='total_bill',
    y='tip',
    color = 'hotpink',
    linestyle = '--',
    linewidth = 3,
    marker = 'o',
    markerfacecolor = 'indigo')
tipgraph.set(title='Tips vs. Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot
tipgraph = sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    style='time',
    hue='day',
    palette=['purple', 'hotpink', 'deepskyblue', 'yellowgreen']
)

# Set title and axis labels
tipgraph.set(
    title='Tips vs. Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)'
)

# Show the plot
plt.show()
```

```python
sns.pairplot(data = tips,
             hue = 'day')
```

```python
daysplot = sns.relplot(data = tips,
    x= 'total_bill',
    y= 'tip',
    hue= 'sex',
    col= 'day',
    kind= 'scatter',
    col_wrap=2)

daysplot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Relational plot with multiple columns
daysplot = sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    col="day",
    kind="scatter",
    col_wrap=2
)

# Adjust the layout for better readability
daysplot.set_axis_labels("Total Bill ($)", "Tip Amount ($)")
daysplot.fig.suptitle("Tips vs. Total Bill by Day and Sex", y=1.02)

# Show the plot
plt.show()
```

```python
import plotly.graph_objects as go # 'go' is 'graph objects

x1 = np.array(['Luffy',
'Zoro',
'Nami',
'Usopp',
'Sanji'])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
graph = go.Figure()
graph.add_trace(go.Bar(x =x1, y =y1))

graph.update_layout(title = 'Pirate Scores',
                    xaxis_title = 'Pirates',
                    yaxis_title = 'Score')
```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(x=x1, y=y1, mode='markers'
, # we want points for a scatter plot
marker=dict(
size=15, # point size
color='hotpink'
, # point colour
opacity=1, # point transparency/alpha
line=dict(width=5, color='purple') # point outline
)))
graph.update_layout(
title='Interactive Pirate Plot'
,
xaxis_title='Pirates'
,
yaxis_title='Scores'
,
width=500, height=500)
```

```python
from wordcloud import WordCloud
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
on_bad_lines='skip')
df
```

```python
# join all our text from each row from our quote column into a string
text = " ".join(each for each in df.quote)
# generate our wordcloud image
wordcloud = WordCloud(background_color="white",
colormap = 'inferno').generate(text)
# use matplotlib syntax to put our image in a figure
fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, # remember 'imshow' from when we added pictures to our matplotlib axes
interpolation='bilinear') # this line helps smooth our image
ax.axis("off")
plt.show()
```

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])
```

```python
diagram = venn2_unweighted([A, B],
    set_labels = ('Fruits', 'Pies'),
    set_colors=("blue", "red"),
    alpha=0.5)
plt.show()
```

```python
diagram = venn2_unweighted([A, B],
    set_labels = ('Fruits', 'Pies'),
    set_colors=("blue", "red"),
    alpha=0.5)

diagram.get_label_by_id('10').set_text("\n".join(A - B))  # Elements only in A
diagram.get_label_by_id('11').set_text("\n".join(A & B))  # Elements in both A and B
diagram.get_label_by_id('01').set_text("\n".join(B - A))
```
