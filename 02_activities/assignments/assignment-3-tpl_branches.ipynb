### Creating Visual for Total Services Available at TPL Branches 

```python
# load data to df
tpl_branches = pd.read_csv("C:\\Users\\Localadmin\\DSI_directory\\visualization\\02_activities\\assignments\\tpl-branch-general-information-2023.csv")

# display first 10 rows of the dataframe
tpl_branches.head(10)

# display column names of the dataframe
tpl_branches.info()
```

```python
# change the format of values in columns - first fill NaN values with 0, then convert to int
tpl_branches[['KidsStop','LeadingReading','CLC','DIH','TeenCouncil','YouthHub','AdultLiteracyProgram','Workstations']].fillna(0).astype(int)
```

```python
# Creating Visual for TPL Branches focused on Digital Innovation Hub
fig, ax = plt.subplots(figsize=(10, 7))

services = tpl_branches[['KidsStop','LeadingReading','CLC','DIH','TeenCouncil','YouthHub','AdultLiteracyProgram']].fillna(0).sum()

bars = ax.barh(services.index, services.values, 
       color='skyblue',
       edgecolor='black')

ax.set_ylabel('Services')

ax.set_xlabel('Number of Branches Offering Service')

ax.set_title('Total Available Services at Toronto Public Library Branches')

plt.xticks(ha='center', wrap=True)

# Add labels on top of each bar
ax.bar_label(bars, fmt='%d', padding=3)

#sort bars in descending order


plt.tight_layout()
plt.show()
```
