```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.lineplot(
	data=tips,        # choose our dataset
	x='total_bill',   # define our x variable
	y='tip'           # define our y variable
)
```

```python
tipgraph = sns.lineplot(
	data=tips,
	x='total_bill',
	y='tip'
)
tipgraph.set(
	title='Tips vs. Total Bill',
	xlabel='Total Bill ($)',
	ylabel='Tip Amount ($)'
)
```

```python
tipgraph = sns.lineplot(
	data=tips,
	x='total_bill',
	y='tip',
	color='hotpink',
	linestyle='--',
	linewidth=3,
	marker='o',
	markerfacecolor='indigo'
)
```

```python
tipgraph = sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    style='time',
    hue='day',
    palette=['purple', 'hotpink', 'deepskyblue', 'yellowgreen']
)
tipgraph.set(
    title='Tips vs. Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)'
)
```

```python
sns.pairplot(data=tips, hue='day')
```

```python
# Create a relational scatter plot of tips vs. total bill, colored by sex, separated by day, with 2 columns per row
daysplot = sns.relplot(
    data=tips,         # Use the 'tips' DataFrame as the data source
    x="total_bill",    # Set 'total_bill' as the x-axis variable
    y="tip",           # Set 'tip' as the y-axis variable
    hue="sex",         # Color the points by the 'sex' column
    col="day",         # Create separate plots for each day of the week
    kind="scatter",    # Use scatter plot as the plot type
    col_wrap=2         # Wrap the columns so there are 2 plots per row
)
```

```python
import plotly.graph_objects as go

x1 = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
graph = go.Figure()
graph.add_trace(go.Bar(x=x1, y=y1))
graph.update_layout(
	title="Pirate Scores",
	xaxis_title="Pirates",
	yaxis_title="Score"
)
graph.show()
```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(
	x=x1,
	y=y1,
	mode='markers',
	marker=dict(
		size=15,
		color='hotpink',
		opacity=1,
		line=dict(width=5, color='purple')
	)
))
graph.update_layout(
	title='Pirate Scores',
	xaxis_title='Pirates',
	yaxis_title='Score',
	width=500,
	height=500
)
graph.show()
```

```python
# Read the CSV file from the URL, skipping bad lines if any
df = pd.read_csv(
	"https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
	on_bad_lines='skip'
)
df
```

```python
# join all our text from each row from our quote column into a string
text = " ".join(str(each) for each in df.quote)

# generate our wordcloud image
wordcloud = WordCloud(background_color='white', colormap='inferno').generate(text)

# use matplotlib syntax to put our image in a figure
fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
# remember 'imshow' from when we added pictures to our matplotlib axes
# this line helps smooth our image
```

```python
# Install matplotlib-venn if not already installed
%pip install matplotlib-venn

from matplotlib_venn import venn2, venn2_circles

A = set(["apple"])
B = set(["pumpkin", "banana", "blueberry", "watermelon", "apple", "key lime"])
diagram = venn2(
	[A, B],
	set_labels=('Fruits', 'Pies'),
	set_colors=("blue", "red"),
	alpha=0.5
)

# Set the labels for each region of the Venn diagram
diagram.get_label_by_id("10").set_text("\n".join(A - B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B - A))


plt.show()
```
