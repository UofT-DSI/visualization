```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
import numpy as np

y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x2, y2, color='red')

#ax.errorbar(x2, y2, 
#            yerr=y2_sd,
#            fmt='none')


ax.errorbar(x2, y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor = 'indigo',
            elinewidth = 4,
            capsize = 6,
            capthick = 4)
```

```python
from PIL import Image
import requests
from io import BytesIO

response = requests.get ('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response. content)
image = Image.open(image_file)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x2, y2, color='red')

ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])
ax_image.imshow(image)
ax_image.axis('off')
```
