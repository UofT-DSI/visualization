Making a basic figure with matplotlib

```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#make some sample data
np.random.seed(613)
x = np.arange(50) #x axis is a range between 1 and 50 (really 0 and 49)
y = np.random.randint(0, 100, 50) # y axis is a random integer between 0 and 100 with 50 values
```

```python
# plot the generated data in a scatterplot
fig, ax = plt.subplots(figsize = (5, 3)) # make a figure "fig", add an axes into that figure. initiate figure with size 5 wide x 3 height
ax.scatter(x,y) # the axes object is going to be a scatterplot with x and y data
```

```python
# we can now modify the figure
# edit axes object and make a bar graph
fig, ax = plt.subplots(figsize = (5,3))
ax.bar(x,y)
```

```python
#to make a horizontal bars
fig, ax = plt.subplots(figsize = (5,3))
ax.barh(x,y)
```

```python
# make a line graph
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)
```

```python
# try it: create a histogram
fig, ax = plt.subplots(figsize = (5,3))
ax.hist(y) # it only counts y values so we dont need x
```

Add plot labels and titles

```python
# make plot first
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

#adding axis labels and titles
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout() # make sure all axes and labels fit into the graph output, dependent on how long they are # to squish subplots closer together
```

```python
# modify labels and titles
# we use font dictionaries ('fontdict')
font1 = {'family': 'sans-serif', 'color':'blue', 'size' : 20}
font2 = {'family': 'monospace', 'color': 'green', 'size' : 14}
# it doesnt have output, we just defined some variables
```

```python
# assign those fonts to our graph
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growht', fontdict= font2)
ax.set_xlabel('Years since start', fontdict= font2)
```

Change alignment of labels

```python
#using location argument loc
# to left align title
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growht', fontdict= font2)
ax.set_xlabel('Years since start', fontdict= font2)
fig.tight_layout() # to squish subplots closer together
```

Modify data points

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,
           y,
           marker = '*', # can be triangles, arrows, etc
           color = 'indigo')
fig.show()
```

Change line style and width

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,
        y,
        marker = '*',
        color = 'indigo', # can also use hexcodes
        linestyle = '--',
        linewidth = 2)
fig.show()
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,
        y,
        marker = '*',
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2)
fig.show()
```

Modifying data points

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,
        y,
        marker = '*',
        markersize=12, #NEW
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359', #NEW
        markerfacecolor = '#000000') #NEW
```

Grid lines

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,
        y,
        marker = '*',
        markersize=12, 
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359', 
        markerfacecolor = '#000000') 
ax.grid() #NEW
```

```python
#modify color, style, width of gridlines
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,
        y,
        marker = '*',
        markersize=12, 
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359', 
        markerfacecolor = '#000000') 
ax.grid(axis = 'y', #define which grid axis you want to change
        color = 'blue',
        linestyle = '-.',
        linewidth = 2)
```

```python
import seaborn as sns

#load dataset
df = sns.load_dataset('iris')

#plot elements
fig, ax = plt.subplots(figsize = (5,3))
sns.set_theme(style="darkgrid")
sns.kdeplot(df['sepal_width'], fill=True, color="r")
sns.kdeplot(df['sepal_length'], fill=True, color="b")
#plt.show()


#adding axis labels and titles
ax.set_title('Sepal width and length')
ax.set_ylabel('Sepal length')
ax.set_xlabel('Sepal width')
fig.tight_layout()

ax.grid(axis = 'y', #define which grid axis you want to change
        color = 'gray',
        linestyle = '-',
        linewidth = .5)
```

aesthetic - visualization has a good contrast of colours, good font size and is not too crowded. <br>
substantive - the visualization represents well the differences between colours, the scale is easy to understand <br>
perceptual - it does not have a legend indicating what each colour refers to, so it cannot be accurately interpretedand draw conclusions from it
