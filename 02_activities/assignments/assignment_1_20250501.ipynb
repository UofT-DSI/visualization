Subplots and combining visualizations

```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# generate sample data
np.random.seed(631)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# make plot first
fig, (ax1, ax2) = plt.subplots(ncols = 2,
                               nrows = 1,
                               figsize = (7, 3))
ax1.scatter(x1, y1,
            marker = "+",
            color = 'red')
ax2.bar(x2, y2,
        color = 'green')
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize = (7,4))
someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1, y1)
```

```python
#modify figure layout
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize = (7,4))
someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1, y1)

someaxes['ax1'].set_xlabel('A big label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18)
```

```python
# fixing label crowding with layouts
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize = (7,4),
                                    layout = 'constrained')
someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1, y1)

someaxes['ax1'].set_xlabel('A big label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18)
```

```python
# make new sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

fig, ax = plt.subplots(figsize = (7,4))

# add now the plot elements
ax.bar(x, y1,
       color = 'indigo')
ax.plot(x, y2,
        color = 'red')
```

```python
# add a new element: error information
# calculate sd
y2_sd = np.std(y2)

# now plot line
fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color = 'red')
```

```python
# add sd
fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color = 'red')

# now add sd
ax.errorbar(x, #our x values
            y2, #our y values
            yerr = y2_sd, # error information
            fmt = "none")
```

```python
fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color = 'red')
ax.errorbar(x,
            y2,
            yerr = y2_sd, 
            fmt = "none",
            ecolor = 'indigo', 
            elinewidth = 4,
            capsize = 6,
            capthick = 4)
```

```python
fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color = 'red')
ax.errorbar(x,
            y2,
            yerr = y2_sd, 
            fmt = "none",
            ecolor = 'indigo', 
            elinewidth = 4,
            capsize = 6,
            capthick = 4,
            errorevery = 2)
```

```python
# ADDING IMAGES
# load extra libraries
from PIL import Image
import requests
from io import BytesIO
```

```python
# get image from internet
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
# make basic plot
fig, ax = plt.subplots(figsize = (7, 3))
ax.plot(x, y2, color = 'red')
```

```python
# overlay new axis on top of existing in the same container
fig, ax = plt.subplots(figsize = (7, 3))
ax.plot(x, y2, color = 'red')

# create a new box where the image will go
ax_image = fig.add_axes([0.1, # x coordinate of image
                        0.11, # y coordinate of image
                        0.15, # width of image
                        0.35]) # height of image
```

```python
fig, ax = plt.subplots(figsize = (7, 3))
ax.plot(x, y2, color = 'red')

ax_image = fig.add_axes([0.1, 
                        0.11, 
                        0.15, 
                        0.35])

# add image
ax_image.imshow(image)
ax_image.axis('off')
```
