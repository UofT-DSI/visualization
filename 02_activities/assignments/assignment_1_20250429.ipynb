Customizing plots

```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# create sample data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
# let's try making a line plot using both y variables in the same axes 
# now we make a line graph
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, y2)
```

```python
# add a legend
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right')
```

```python
# modify a legend
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
          frameon = True, #adds a frame around the legend
          fontsize = 12,
          ncol = 2, #number of columns
          shadow = True) # adds a drop shadow to left hand
```

```python
# move a legend outside of plot
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left', # this refers to upper left corner of the legend. use that as anchor point
          bbox_to_anchor = (1,1)) # i want my legend to sit horizontally 100% - meaning the rightmost side of horizontal axes. and 100% on vertical axes, in the upper right corner
```

```python
# make a scatter plot
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
```

```python
# add anotation using text() function
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important")
```

```python
# modify text annotation
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important!",
        ha = 'center', # horizontal alignment of text
        color = 'red', # modify font colour
        size = 20)
```

```python
# positioning text annotation
fig, ax = plt.subplots()
ax.axis([0,10,0,10])
```

```python
fig, ax = plt.subplots()
ax.axis([0,10,0,10])

ax.text(1, 5, ". Data:(1,5)", transform = ax.transData)
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform = ax.transAxes)
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform = fig.transFigure)
```

```python
# annotating with arrows to point exactly to something
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', 
            xy = (10,95), # where we want our arrow to point
            xytext = (20, 94), # where we want our text
            arrowprops = dict(facecolor = 'black',
                              shrink = 0.05)
                              )
```

```python
# modify arrows to point exactly to something
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', 
            xy = (10,95), # where we want our arrow to point
            xytext = (20, 94), # where we want our text
            arrowprops = dict(arrowstyle = 'wedge',
                              color = 'hotpink')
                              )
```

```python
# remove ticks marks and labels
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())

#no labels and ticks in y axis, no labels on x axis
```

```python
# limit ticks marks
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #set maximum number of ticks on x axis
```

```python
# tick mark intervals
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# rotating axis labels
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation = 45, ha = 'center')
```

```python
# activity - modify axis title fonts
# make serif font, color indigo, name of axis: 'shiny new x axis' (add axis label)
font_color = {'family' : 'serif', 'color' : 'indigo'} #create dictionary with elements we want
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xlabel('Shiny New X-Axis', fontsize = 18, fontdict = font_color) # we call out our dictionary here
```
