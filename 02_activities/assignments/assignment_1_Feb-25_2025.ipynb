```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(613) # set seed
x = np.arange(50) # 0,1,...,49
y = np.random.randint(0,100,50) # 50 random integers with min = 0, max = 99
fig,ax = plt.subplots(figsize=(5,3)) # figure size
ax.scatter(x,y)
ax.bar(x,y)
ax.plot(x,y)
ax.hist(y)
fig,ax = plt.subplots(figsize=(5,3)) 
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
font1 = {'family':'sans-serif', 'color':'blue', 'size':20}
font2 = {'family':'monospace', 'color':'green', 'size':14}

np.random.seed(613) # set seed
x = np.arange(50) # 0,1,...,49
y = np.random.randint(0,100,50) # 50 random integers with min = 0, max = 99
fig,ax = plt.subplots(figsize=(5,3)) # figure size
ax.scatter(x,y)
ax.bar(x,y)
ax.plot(x,y)
ax.hist(y)
fig,ax = plt.subplots(figsize=(5,3)) 
ax.plot(x,y)

ax.set_title('Total growth over time',fontdict = font1)
ax.set_ylabel('Total growth',fontdict = font2)
ax.set_xlabel('Years since start',fontdict = font2)
plt.show()
```

```python
#3.Recall the aestetics, substintive and perceptual qualities of data visualization.
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

fig, ax = plt.subplots()


n, bins, patches = ax.hist(data, bins=10, edgecolor='black')

# Calculate the index of the bar with the maximum value
max_index = np.argmax(n)

# Assign different colors to each bar and highlight the max value in red
for i, patch in enumerate(patches):
    if i == max_index:
        patch.set_facecolor('red')
    else:
        patch.set_facecolor(plt.cm.viridis(i / len(patches)))

# Add titles and labels
font1 = {'family':'sans-serif', 'color':'indigo', 'size':15}
font2 = {'family':'monospace', 'color':'indigo', 'size':12}

ax.set_title('Histogram with Max Value Highlighted in Red',fontdict = font1, loc = 'center')
ax.set_xlabel('Value',fontdict = font2)
ax.set_ylabel('Frequency',fontdict = font2)

plt.show()
```
