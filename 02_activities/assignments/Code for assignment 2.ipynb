```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Adry","Meraz","Rivera","Star","Canada"])
y2 = np.array([110,18,240,99,220])

font3 = {'family':'serif', 'size':16}
font2 = {'family':'serif', 'size':16}
    
fig,(ax1,ax2) = plt.subplots(ncols=2,
                             nrows=1,
                             figsize=(7,3))
ax1.scatter(x1,y1,
            marker = 'o',
            color = 'red')
ax2.bar(x2,y2, color = 'darkgreen')

ax1.set_title('Scatterplot', fontdict = font3)
ax1.set_ylabel("Measure",fontdict = font2)
ax1.set_xlabel('Time',fontdict= font2)
```

```python
# specifying sport we give to each chart. we could have added ax1,ax1  and ax2,ax3 and that will show the first chart horixontal and the other two at the bottom
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'], 
                                   ['ax2','ax3']],
                                   figsize = (7,4))

someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furiois', fontsize = 18)
```
