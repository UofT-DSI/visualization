Using the Heart Failure Prediction dataset from Kaggle:
https://www.kaggle.com/datasets/fedesoriano/heart-failure-prediction

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
```

```python

# Load the dataset
heart = pd.read_csv('heart.csv', on_bad_lines='skip')  # Adjust the path if needed
print(heart.info())

# Drop rows with any missing values
heart_clean = heart.dropna()
print(heart_clean.info())


# Randomly sample 50 rows
heart_sample = heart_clean.sample(n=50, random_state=42)  # random_state set for reproducibility
heart_sample
```

```python
# Set the style of the plot to dark grid
sns.set_style("darkgrid")

# Create a bar plot using seaborn
# x-axis: HeartDisease (0/1)
# y-axis: Age
# Color bars based on HeartDisease value
tipgraph = sns.barplot(data=heart,
                           x = 'HeartDisease',
                           y = 'Age',
                           hue='HeartDisease',
                           legend=False,
                           )

# Set the title and labels for the plot
tipgraph.set(title = 'Age vs Heart Disease',
             xlabel = 'Heart Disease (0 = No, 1 = Yes)',
             ylabel = 'Age')
```
