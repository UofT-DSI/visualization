Using the Heart Failure Prediction dataset from Kaggle:
https://www.kaggle.com/datasets/fedesoriano/heart-failure-prediction

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
```

```python

# Load the dataset
heart = pd.read_csv('heart.csv', on_bad_lines='skip')  # Adjust the path if needed

# Drop rows with any missing values
heart_clean = heart.dropna()

# Randomly sample 50 rows
heart_sample = heart_clean.sample(n=50, random_state=42)  # random_state set for reproducibility
```

```python
sns.set_style("darkgrid")
tipgraph = sns.barplot(data=heart,
                           x = 'HeartDisease',
                           y = 'Age',
                           palette='Set1')

tipgraph.set(title = 'Age vs Heart Disease',
             xlabel = 'Heart Disease (0 = No, 1 = Yes)',
             ylabel = 'Age')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load and clean data
df = pd.read_csv('heart.csv')
df_clean = df.dropna()

# Plot cholesterol vs heart disease
sns.barplot(x='HeartDisease', y='Cholesterol', data=df_clean)
plt.title('Cholesterol Levels by Heart Disease Status')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Cholesterol (mg/dL)')
plt.show()
```
