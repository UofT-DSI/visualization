```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
import seaborn as sns
import plotly.graph_objects as go

```

```python
# Load and display the data
df = pd.read_csv("W:/DSI/visualization/02_activities/assignments/assignment_3/wellbeing_toronto.csv")

df
```

```python
# Make a scatterplot: Healthy Food Index vs. EDI Score Across Neighbourhood
fig = sns.scatterplot(data=df, x='Early Development Instrument (EDI)', y='Healthy Food Index', hue='Neighbourhood')

fig.set (title='Healthy Food Index vs. Early Development Instrument Score by Toronto Neighbourhood',
             xlabel='Early Development Instrument (EDI)',
             ylabel='Healthy Food Index')

# Move legend below the plot
plt.legend(
    title='Neighbourhood',
    bbox_to_anchor=(0.5, -0.25),  # Center below plot
    loc='upper center',
    ncol=3,  # Adjust for how many columns of legend entries
    frameon=False
)
```

```python
## Make an interactive version
# Prepare x, y, and hover text
x1 = df['Early Development Instrument (EDI)']
y1 = df['Healthy Food Index']
hover_text = df.apply(lambda row: f"Neighbourhood: {row['Neighbourhood']}<br>"
                                  f"EDI Score: {row['Early Development Instrument (EDI)']:.2f}<br>"
                                  f"Healthy Food Index: {row['Healthy Food Index']:.2f}", axis=1)


# Create the scatter plot
graph = go.Figure()
graph.add_trace(go.Scatter(
    x=x1,
    y=y1,
    mode='markers',
    text=hover_text,
    hoverinfo='text',
    marker=dict(
        size=12,
        color='hotpink',
        opacity=0.8,
        line=dict(width=1.5, color='purple')
    )
))

# Update layout with titles and size
graph.update_layout(
    title='Healthy Food Index vs. Early Development Instrument Score by Toronto Neighbourhood (Interactive)',
    xaxis_title='Early Development Instrument Score (EDI)',
    yaxis_title='Healthy Food Index',
    width=800,
    height=600
)

graph.show()
```
