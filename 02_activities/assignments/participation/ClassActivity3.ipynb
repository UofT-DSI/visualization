```python
# load libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# Sample data

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(['Luffy','Zoro','Nami','Ssopp','Sanji'])
y2 = np.array([110,180,240,99,220])
```

```python
# Creating two plots in one figure
fig, (ax1, ax2) = plt.subplots(ncols =2, # number of columns = side by side
                               nrows = 1, # number of rows
                               figsize=(7,3)) # size of figure
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
fig.show()
```

```python
# Activity: Customizing plots
plt.style.use('ggplot')
fig, (ax1, ax2) = plt.subplots(ncols =2, # number of columns = side by side
                               nrows = 1, # number of rows
                               figsize=(7,3)) # size of figure
ax1.scatter(x1,y1, color ='hotpink')
ax2.bar(x2,y2, color = "purple")
ax1.set_title('scatter plot')
ax2.set_title('bar plot')
fig.show()
```

```python
# Defining supblots with "moasic"
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                       ['ax2','ax3']],
                                       figsize=(7,4),
                                       layout = 'constrained') # used to make all the labels and graphs fit in the figure
someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

# adding titles to plot
someaxes['ax1'].set_xlabel('A Big Label', fontsize=18)
someaxes['ax2'].set_xlabel('Another Lable', fontsize=18)
someaxes['ax3'].set_xlabel('Lable 2: 2 Fast 2 Furious', fontsize=18)

plt.show()
```

```python
# Defining supblots with "moasic" - without specifying the layout
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                       ['ax2','ax3']],
                                       figsize=(7,4)
                                        ) 
someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

# adding titles to plot
someaxes['ax1'].set_xlabel('A Big Label', fontsize=18)
someaxes['ax2'].set_xlabel('Another Lable', fontsize=18)
someaxes['ax3'].set_xlabel('Lable 2: 2 Fast 2 Furious', fontsize=18)
```

```python
plt.style.use('default')
# Multiple viz on one axes object

# Sample data
x = np.array(['Luffy','Zoro','Nami','Usopp','Sanji'])
y1 = np.array([110,180,240,99,220])
y2=np.array([170,100,90,120,50])

fig, ax = plt.subplots(figsize=(7,3))

ax.bar(x,y1, color = 'indigo')
ax.plot(x,y2, color = 'red')

fig.show()
```

```python

y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x,y2, color='red')

ax.errorbar(x,y2, yerr=y2_sd, fmt='none',
            ecolor='indigo',
            elinewidth=4,
            capsize=6,
            capthick=4)
#plt.margins(x=0.05,y=0.1) # add in margins to sides of graph
```

```python
# Using images from the internet

# load extra libraries

from PIL import Image
from io import BytesIO
```

```python
# define variables to import image from URL
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7,3))

ax.plot(x,y2, color='red')
```

```python
# adding the image to the above axes
fig, ax = plt.subplots(figsize=(7,3))

ax.plot(x,y2, color='red')

# define location of image
ax_image = fig.add_axes([0.1, # x coord (on figure, not the axes)
                        0.11, # y coord
                        0.15, # image width
                        0.35]) # image height

# add the image 
ax_image.imshow(image)
ax_image.axis('off') # removes border around image

fig.show()
```
