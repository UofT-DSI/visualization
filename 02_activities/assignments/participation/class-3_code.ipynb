```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL 
import requests

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50) 
y2 = np.random.randint(0, 100, 50)
```

```python
# make a line plot with both of our y variables on the same axes 

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1) 
ax.plot(x,y2)
fig.show()
```

```python
# lets add a legend using a label argument in the plot function

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1") 
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right')
fig.show()
```

```python
# lets modify our legend 

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1") 
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right', 
        frameon=True,
        fontsize='12',
        ncol =2, #specifcy number of columns
        shadow = True)
fig.show()
```

```python
# move the plot outside the plot area

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
        bbox_to_anchor =(1,1))
fig.show()
```

```python
# annotations with the text() function 

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10,95, "This value is important!")
fig.show()
```

```python
# Modify text annotations 

ax.text(10,95, "This value is important!",
         ha='center',
         color='red',
         size= 20) 

fig.show()
```

```python
# use transforms for positioning text annotations 

fig, ax = plt.subplots() 
ax.axis([0, 10, 0, 10]) 

ax.text(1,5, ". Data:(1,5)",  transform=ax.transData)
ax.text(0.5, 0.1, ".Axes:(0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ".Figure:(0.2, 0.2)", transform=fig.transFigure)
```

```python
# Annotating with arrows 

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10,95), xytext=(20, 94),
          arrowprops=dict(facecolor='black'))

```

```python
# Modifying arrow properties
ax.annotate('This is important!',
            xy=(10,95), xytext=(20, 94),
            arrowprops = dict(arrowstyle= 'wedge',
                            color = 'hotpink'))
fig.show()
```

```python
# removing tick marks and labels 

fig, ax =plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator()) #removes ticks
ax.xaxis.set_major_formatter(plt.NullFormatter()) #removes labels
```

```python
# Define the max number of ticks on each axis 

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) # max 4 ticks on x axis
fig.show()
```

```python
# Tick mark intervals 

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # ticks every 5 units
```

```python
# Rotating axis labels
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha='right')
```

```python
# Modifying axis title fonts 
font1 = {'family':'serif', 'color':'indigo'}
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xlabel('Shiny New X Axis!', fontsize= 18, fontdict= font1)
```
