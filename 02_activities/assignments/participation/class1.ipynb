# Importing Libraries + Creating a simple scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(42)  # Set random seed for reproducibility
x = np.arange(50)  # Generate x values from 0 to 49
y = np.random.randint(0, 100, 50)  # Generate 50 random y values between 0 and 100
```

# Creating varied plot types 

```python
fig, ax = plt.subplots(figsize=(5, 3))  # Create figure and axes
ax.scatter(x, y)  # Generate scatter plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x, y)  # Generate bar plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)  # Generate a line plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y, bins=10)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.set_title("Total Growth Over Time")  # Add title
ax.set_xlabel("Years Since Start")  # Add x-axis label
ax.set_ylabel("Total Growth")  # Add y-axis label
plt.show()
```

```python
font1 = {"family": "serif", "color": "blue", "size": 20}
font2 = {"family": "monospace", "color": "green", "size": 14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.set_title("Total Growth Over Time", fontdict=font1)
ax.set_xlabel("Years Since Start", fontdict=font2)
ax.set_ylabel("Total Growth", fontdict=font2)
plt.show()
```

```python
ax.set_title("Total Growth Over Time", fontdict=font1, loc='left')  # Left align titl
```

```python
#Customizing Marketer Style
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y, marker='*', color='indigo')  # Change marker shape and color
plt.show()
```

```python
## Adding Line Styles
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker='*', color='indigo', linestyle='--', linewidth=2)  # Dashed line
plt.show()
```

```python
## Using Specific Hex Colors
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker='*', color='#7425B0')
plt.show()
```

```python
## Changing the Marker Edge Colors
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker='o', markersize=12, markeredgecolor='#F9359D', markerfacecolor='#FFC0CB')  # Picked pink as the marker edge and face color
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.grid(True)  # Default grid lines
plt.show()

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.grid(axis='y', color='pink', linestyle=':', linewidth=2)  # Modifying the grid line appearance
plt.show()
```
