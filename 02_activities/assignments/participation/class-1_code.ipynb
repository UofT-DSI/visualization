```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# make some sample data 
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)

# make basic scatterplot 
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
# lets modify the graph 

fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
# lets make a histogram 

fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y) 
ax.hist(y, edgecolor='orange')
```

```python
# adding axis labels 
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
ax.set_title("Total Growth Over Time")
ax.set_ylabel("Total Growth")
ax.set_xlabel("Years since start")
fig.tight_layout()
```

```python
# Modifying labels and titles 
font1 = {'family': 'sans-serif', 'color':  'blue', 'size': 16 } 
font2 = {'family': 'monospace', 'color': 'green', 'size': 14 }
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title("Total Growth Over Time", fontdict=font1)
ax.set_ylabel("Total Growth", fontdict=font2)
ax.set_xlabel("Years since start", fontdict=font2)
fig.tight_layout()
```

```python
# customize plot apperance 
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,
           y, 
           marker='*',
           color="indigo")

fig.show()
```

```python
# modify data points 
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,
           y, 
           marker='*',
           color="indigo", 
           linestyle='--',
           linewidth=2,)

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

# color being the new addition 
ax.plot(x,y,marker='*',color ='#7425b9', linestyle='--', linewidth=2)
fig.show()
```

```python
# customize the plot further 
ax.plot(x,y,marker='*',
        color ='#7425b9', 
        linestyle='--', 
        linewidth=2,
        markeredgecolor='fa9359',
        markerfacecolor='#000000')
fig.show() 

ax.grid(axis ='y', 
        color = 'red',
        linestyle = '--',
        linewidth = .05)
```
