```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
# make a line plot with both y variables on same axes

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

## ADDING A LEGEND

```python
# To add a legend:
# 1. add a label argument to each plot object
# 2. call "legend" and specify its location

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
fig.show()
```

```python
# MODIFYING A LEGEND   

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
        frameon = True,   #add frame around the legend
        fontsize = 12,    #change font size
        ncol = 2,         #specify number of columns
        shadow = True )
fig.show()
```

```python
# MOVING THE LEGEND OUTSIDE PLOT AREA

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
        bbox_to_anchor =(1, 1))
fig.show()
```

## TEXT AND ANNOTATIONS

```python
# Annotations with the text() function

# 1. Put data into a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# 2. use the text() function to specify coordinates and text
ax.text(10, 95, "This value is important!")
fig.show()
```

```python
# Modifying text annotations: add arguments

ax.text(10, 95, "This value is important!",
      ha = 'center',    
      color = 'red',  
      size = 20)      
fig.show()


## this is supposed to print "The value is important" in large red letters 
```

```python
# Positioning text annotations
# use transforms
# start by making a simple,empty set of axes
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
```

```python
# then position the annotation with reference to the data, the axes or the figure
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
ax.text(1, 5, ". Data:(1, 5)",           transform=ax.transData    )
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)",   transform=ax.transAxes    )
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure )
```

```python
# Annotating with arrows

# use annotate()
# must specify location to point at and the location where we want the text
 
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
          arrowprops=dict(facecolor='blue'))
fig.show()
```

```python
# Modifying arrows

# use arrowprops arg
# see (https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.annotate.html)
# try "->","<->","-[","fancy"

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
          arrowprops=dict(facecolor='hotpink', arrowstyle = "wedge"))
fig.show()
```

## AXIS LABELS

```python
# Removing tick marks and labels

# use locator to remove ticks and labels
# use formatter to remove only labels

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# Defining number of tick marks

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# tick mark intervals. e.g., 5

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# Rotating axis labels

# use xticks() to rotate axis labels and horizontal alignment

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha='right', fontsize=12, font = "serif")
```

## SUBPLOTS

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)        
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# so far we've done:

fig, ax = plt.subplots(figsize=(5,3))
```

```python
# now we want 2 plots next to each other, so need to define multiple axes and their relative positions:

fig,(ax1, ax2) = plt.subplots(ncols=2,
                        nrows=1,
                        figsize=(7,3))
```

```python
# then we use our new figure with 2 axes, and define the types of visualizations we want to see in each:

fig,(ax1, ax2) = plt.subplots(ncols = 2,
                            nrows = 1,
                            figsize=(7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
fig.show()
```

```python
# Subplots without a grid arrangement
# use plt.subplot_mosaic()

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))
```

```python
# Subplots without a grid arrangement - with data

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

## MODIFY FIGURE LAYOUT

```python
# Layouts
# Add some very large x axis titles

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```

# Labels are missing or cut off. To fix, can use 1 of 2 types of layouts
# 1. Tight - adjusts subplots so ticks, axis labels, titles don't overlap
# 2. Constrained - also fits things like legends or colorbars

```python
# constrained:

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4),
                                layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)

```

## MULITPLE VIZ ON ONE AXES. (LAYERING PLOTS)

```python
# On one axes object:  just call multiple plot methods on the same axes
# annotaions, shapes, etc can be added as usual

# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))

# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1, color = "indigo")
ax.plot(x, y2, color = "red")
```

## LAYERING ERROR INFORMATION

```python
# Add error info

y2_sd = np.std(y2)
```

```python
# then plot line as before:

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
```

```python
# then use errorbar() to add error line (std in this case)
# yerr = plotting vertical error bars
# fmt (format) makes sure we're not plotting the actual data points, only the error
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax.errorbar(x, #our x values
        y2, #our y values
        yerr = y2_sd,
        fmt = "none")
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
        color = "red")
ax.errorbar(x,y2,
        yerr = y2_sd,
        fmt = "none",
        ecolor= "indigo",
        elinewidth= 4,
        capsize = 6,
        capthick= 4 )
```

```python
# to not see error bars for every single point, can specify intervals using errorevery (probably no use case for this)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
        color = "red")
ax.errorbar(x,
        y2,
        yerr = y2_sd,
        fmt = "none",
        ecolor= "indigo",
        elinewidth= 4,
        capsize = 6,
        capthick= 4,
        errorevery= 2)
```

## IMAGES

```python
# 1. load libraries

from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images
```

```python
# 2. get image from the internet

response = requests.get('https://static.wikia.nocookie.net/the-ultimate-good/images/3/3d/Luffy.png/revision/latest?cb=20230417200546')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
# 3. Add the image to a plot

# i. make a basic line plot 
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
```

```python
# ii. overlay a new axis (ax_image) on our figure (on top of ax) to act as a container for our image

ax_image = fig.add_axes([0.1, # x coordinate (ON FIGURE, NOT AXES)
                        0.11, # y coordinate (ON FIGURE, NOT AXES)
                        0.15, # image width
                        0.35]) # image height
```

```python
# iii. add imshow() to add the image we prepped

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2,
        color="red")
ax_image=fig.add_axes([0.1, 0.11, 0.15, 0.55])
ax_image.imshow(image)
ax_image.axis('off')
```
