```python
# load libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# Sample data

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
# Create a line plot using both y variables on the same axes

fig, ax = plt.subplots(figsize=(5,3)) 
ax.plot(x,y1)
ax.plot(x,y2)
fig.show() 
```

```python
fig, ax = plt.subplots(figsize=(5,3)) 
ax.plot(x,y1, label ='Person 1')
ax.plot(x,y2, label='Person 2')
ax.legend(loc='lower right')
fig.show() 
```

```python
fig, ax = plt.subplots(figsize=(5,3)) 
ax.plot(x,y1, label ='Person 1')
ax.plot(x,y2, label='Person 2')
ax.legend(loc='lower right',
          frameon= True, # adds border around the legend
          fontsize = 12,
          ncol = 2, # number of columns
          shadow = False)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label="Person 1")
ax.plot(x, y2, label="Person 2")
ax.legend(loc = 'upper left',
          bbox_to_anchor = (1,0.3)) # (x,y) 
fig.show()
```

```python
# Annotations with text

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label='Person 1')
ax.scatter(x,y2, label="Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important", # the point defined is at the beginning of the text
        ha = "center", # when we specific the HA to center, the point is now at the center of the  text
        color = 'red',
        size = 12) 
fig.show()
```

```python
# Positioning text annotations

fig, ax = plt.subplots()
ax.axis([0,10,0,10])

ax.text(1,5, ". Data: (1,5)", transform=ax.transData)
ax.text(0.5,0.1, ". Axes: (0.5,0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform=fig.transFigure)
```

```python
# Annotating with arrows
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label='Person 1')
ax.scatter(x,y2, label="Person 2")
ax.legend(loc = "lower right")

ax.annotate('This is important', xy=(10,95), xytext=(20,94),
           arrowprops=dict(facecolor = "black", shrink = 0.05))
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label='Person 1')
ax.scatter(x,y2, label="Person 2")
ax.legend(loc = "lower right")

ax.annotate('This is important', xy=(10,95), xytext=(20,94),
           arrowprops=dict(arrowstyle = 'wedge', color ='hotpink'))
```

```python
# Removing tick marks and labels

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label='Person 1')
ax.scatter(x,y2, label="Person 2")
ax.legend(loc = "lower right")

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# Tick limiter
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label='Person 1')
ax.scatter(x,y2, label="Person 2")
ax.legend(loc = "lower right")

ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# Tick matk interva;s
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label='Person 1')
ax.scatter(x,y2, label="Person 2")
ax.legend(loc = "lower right")

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# rotating tick marks
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label='Person 1')
ax.scatter(x,y2, label="Person 2")
ax.legend(loc = "lower right")

plt.xticks(rotation=45, ha ='center')
```
