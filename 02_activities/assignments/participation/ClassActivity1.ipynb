```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50) # range of 0 to 49
y = np.random.randint(0,100,50) # random values between 0 and 100, 50 values
```

```python
fig, ax = plt.subplots(figsize=(5,3)) # set ax object and put it in a "container" aka fig
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.barh(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y) # just need y-values for histgrams 
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title("Total growth over time")
ax.set_ylabel("Total growth")
ax.set_xlabel("Years since start")
fig.tight_layout() # ensures all the lables fit into the graph/figure
```

```python
# changine fonts of lables

# first we create font dictionaries to specify the font type, colour, and size
font1 ={'family':'sans-serif','color':'blue', 'size':20}
font2 = {'family':'monospace', 'color':'green', 'size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title("Total growth over time",fontdict=font1, loc ='left') # use loc to change alignment
ax.set_ylabel("Total growth", fontdict= font2)
ax.set_xlabel("Years since start", fontdict=font2)
fig.tight_layout() # ensures all the lables fit into the graph/figure
```

```python
# Modifying data points
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y, 
           marker ='*',
           color = 'indigo')

```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y, 
           marker ='*',
           color = 'indigo',
           linestyle ='--',
           linewidth= 2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y, 
           marker ='*',
           color = '#7425b9', # can also modify colour by hex codes
           linestyle ='--',
           linewidth= 2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y, 
           marker ='*',
           markersize= 12,
           color = '#7425b9', # can also modify colour by hex codes
           linestyle ='--',
           linewidth= 2,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')
```

```python
# Adding grid lines
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y, 
           marker ='*',
           markersize= 12,
           color = '#7425b9', # can also modify colour by hex codes
           linestyle ='--',
           linewidth= 2,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')
ax.grid(axis ='y',
        linestyle ='--',
        linewidth = 2,
        color = 'blue')
```
