```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
np.random.seed(613) # Setting the random seed for reproducibility
x = np.arange(50) # Generating the x-coordinates (values from 0 to 49)
y = np.random.randint(0,100,50) # Generating the y-coordinates (50 random integers between 0 and 100)

fig,ax = plt.subplots(figsize = (5,3)) # Creating the figure and axis for the plot
# figsize=(5, 3) specifies the width and height of the plot in inches
ax.scatter(x,y) # Creating the scatter plot with x and y coordinates
```

```python
fig, ax = plt.subplots(figsize= (5,3))
ax.bar(x,y) # Generating a bar plot
# x: the x-coordinates (categories or positions for the bars)
# y: the heights of the bars (values to be plotted)
```

```python
fig, ax = plt.subplots(figsize= (5,3))
ax.hist(y) # Generating a histogram
# y: the data to be grouped into bins and visualized as a histogram
# By default, the number of bins is automatically determined.
```

```python
fig,ax = plt.subplots(figsize = (5,3))
ax.plot(x,y) # Generating a line plot

ax.set_title ('Total growth over time')# Adding a title to the plot
ax.set_ylabel('Total Growth')# Labeling the y-axis
ax.set_xlabel('Years since start')# Labeling the x-axis
```

```python
# Defining custom font styles for the plot
font1 = {'family':'sans-serif','color':'blue','size': 20}
font2 = {'family':'monospace','color':'green','size':14}

fig,ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)# Generating a line plot

ax.set_title('Total growth over time', fontdict = font1)# fontdict=font1 applies the defined font properties to the title
ax.set_ylabel('Total growth', fontdict = font2)# fontdict=font2 applies the defined font properties to the y-axis label
ax.set_xlabel('Years since start', fontdict= font2)# fontdict=font2 applies the defined font properties to the x-axis label
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y,
           marker = '*', # marker='*': Specifies that the marker for each data point is a star
           color = 'indigo') # color='indigo': Specifies the color of the markers as indigo
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
           marker = 'o', # Marker shape set to a circle ('o')
           color = 'indigo', # Line color set to indigo
           linestyle = '--',  # Line style set to dashed ('--')
           linewidth = 2,  # Line width set to 2
           markeredgecolor = '#F5F5DC', # Edge color of the marker set to beige (hex: #F5F5DC)
           markerfacecolor = '#000000') # Fill color of the marker set to black (hex: #000000)

ax.grid()
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
           marker = 's',
           color = 'indigo',
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')

ax.grid()
```
