```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50) #low, high, size
y2 = np.random.randint(0,100,50)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = 'Person 1')
ax.plot(x,y2, label = 'Person 2')
ax.legend (loc = 'upper left',
           frameon = True,
           fontsize = 12,
        #    ncol = 2,
           shadow = True,           
           bbox_to_anchor = (1,1)
           )
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label='Person 1')
ax.scatter(x,y2,label = 'Person 2')
ax.legend(loc='lower right')
ax.text(10,95, "This value is very important",
        ha = 'center',
        color = 'red',
        size = 16
        )
```

```python
np.random.seed(123)
x1= np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220 ])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(11,3))

ax1.scatter(x1,y1)
ax2.bar(x2, y2, color='darkgreen')

# plt.xlabel("Specific Users")
# plt.ylabel("Amount")

ax1.set_title("Important Scatterplot", fontdict=font1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']
                                    ], figsize=(7,4), layout = "constrained")

someaxes["ax1"].scatter(x1, y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel('A big label', fontsize=18)
someaxes['ax2'].set_xlabel('Another label', fontsize=18)
someaxes['ax3'].set_xlabel('Fast too Furious', fontsize=18)
```

```python
x = np.array(["Luffy", "Zuro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220 ])
y2 = np.array([170, 100, 90, 120, 50 ])

fig, ax = plt.subplots(figsize = (5,3))
ax.bar(x,y1, color = "indigo")
ax.plot(x, y2, color='red')
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color='red')

ax.errorbar(x,y2, 
            yerr=y2_sd, 
            fmt='none', 
            ecolor = 'blue', 
            elinewidth=4, 
            capsize=7, 
            capthick=5, 
            errorevery=2)
```
