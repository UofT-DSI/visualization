```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613)
x = np.arange(50) # give us 50 values
y = np.random.randint(0,100,50) # give us 100 random integers btwn 0 - 50
```

```python
fig, ax = plt.subplots(figsize=(5,3))
# give us a figure (canvas) with the axes (actual plot) as ax
# using plt.subplots = setting the size of our figures in inches

ax.scatter(x,y)
# what we want axes (actual plot) to look like - what kind of data we want it to plot
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
# exact same data, exact same size; the only thing we need to change is the function ax.bar
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.barh(x,y) # horizontal bar graphs
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y) #histogram
# why is this only ax.hist(y) and not ax.hist(x,y)?
# x = number of occurences
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
font1 = {'family':'sans-serif', 'color':'blue', 'size':20}
font2 = {'family':'monospace', 'color':'green', 'size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y, 
           marker = '*', 
           color = 'indigo',
           linestyle = '--',
           linewidth = 2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y, 
           marker = '*', 
           markersize = 12,
           color = 'yellow',
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = "#19ac6f",
           markerfacecolor = "#A03DDE")

ax.grid(axis = 'y', color = 'purple', linestyle = '-', linewidth = '3') # if you leave it blank ax.grid(), it shows both x & y 
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd
import numpy as np

data = pd.read_csv('https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/gapminderData.csv')
data['continent'] = pd.Categorical(data['continent'])
data.head()

fig, ax = plt.subplots(figsize=(10, 10))

# Subset of the data for year 1952
data1952 = data[data.year == 1952]

# Scatterplot
ax.scatter(
    x = data1952['lifeExp'], 
    y = data1952['gdpPercap'], 
    s = data1952['pop']/50000, 
    c = data1952['continent'].cat.codes, 
    cmap = "Accent", 
    alpha = 0.6, 
    edgecolors = "white", 
    linewidth = 2
)
 
# Add titles (main and on axis)
plt.yscale('log')
ax.set_xlabel("Life Expectancy")
ax.set_ylabel("GDP per Capita")
ax.set_title("Year 1952")
ax.set_ylim(100,50000)
ax.set_xlim(30, 75)

plt.show()
```
