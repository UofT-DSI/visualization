```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# for x variable give us 50 values, so 0-49
# for y value, we want 50 random integers between 0 and 100
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)
```

```python
# make a basic scatterplot
# we want to make a figure (a canvas) with dimensions 5 x 3 (inches?)
# in that figure we want an axis object. We wnat that axis object to be a scatterplot & we want it to plot x & y


fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
# Now we can start playing with indovidual elements of that graph
# now we want all same elements, except we want a bar graph instead of a scatterplot

fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)

#to make horizontal bar use
ax.barh(x,y)
```

```python
#to make horizontal bar use
fig, ax = plt.subplots(figsize=(5, 3))
ax.barh(x,y)
```

```python
#to make a line plot:

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
# to change the dimensions:

fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(x,y)
```

```python
#to make a histogram: only need the y

fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
# PLOT LABELS

fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
# MODIFYING LABELS AND TITLES
# by using font dictionaries(fontdict)
# Dictionary refresher: in curly brackets, use key:value pairs, e.g., keys =family, color, size; values = sans, blue, 20

# First,make a dictionary with the properties we want for our labels

font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

#then add fontdic arguments to our previous lineplot code

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# MOVING LABELS AND TITLES  
# use the loc argument to Title

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# CUSTOMIZING PLOT APPEARANCE

# MODIFYING DATA POINTS
# this time add color & marker arguments (US spelling)

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,y,
          marker='*',         # NEW ADDITIONS
          color = "indigo")   # NEW ADDITIONS
fig.show()
```

```python
# use + signs

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,y,
          marker='+',         # NEW ADDITIONS
          color = "indigo")   # NEW ADDITIONS
fig.show()
```

```python
# use diamonds signs

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,y,
          marker='D',         # NEW ADDITIONS
          color = "indigo")   # NEW ADDITIONS
fig.show()
```

```python
# can add style & width to line

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,
        y,
        marker='*',
        color = 'indigo',
        linestyle = '--',   # NEW ADDITIONS
        linewidth = 2)      # NEW ADDITIONS
fig.show()
```

```python
# MODIFYING DATA POINTS: COLOUR
# can use named colours. See 
# https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py

# or hex codes

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,
        marker='*',
        markersize = 12,  # NEW ADDITIONS
        color = 'indigo',
        linestyle = '--',   
        linewidth = 2,
        markeredgecolor = '#fa9359',   # NEW ADDITIONS
        markerfacecolor = '#000000')   # NEW ADDITIONS
fig.show()


```

```python
# ACTIVITY

#Code from https://python-graph-gallery.com/all-charts/

#Create dataset
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')

df = pd.DataFrame({'height': height,
                   'bars': bars})


color = ['lightblue']
edgecolor = ['darkblue']

# Create bars
fig, ax = plt.subplots()
ax.bar(df['bars'], df['height'],
       color=color,
       edgecolor=edgecolor,
       linewidth=1)

# add grid lines
ax.grid(axis = 'y', color = 'red', linewidth = 1, linestyle = '-.') 

# Show graph
plt.show()
```
