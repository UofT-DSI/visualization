```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
```

```python
fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7, 3))

# Customize ax1 - scatter plot
ax1.scatter(x1, y1, color='purple', marker='x')
ax1.set_title('Scatter Plot: Observations', fontsize=12, color='darkblue')
ax1.set_xlabel('X Axis', fontsize=10)
ax1.set_ylabel('Y Axis', fontsize=10)
ax1.annotate('Peak Value',
             xy=(x1[y1.argmax()], y1.max()),
             xytext=(x1[y1.argmax()], y1.max() + 1),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=8)

# Customize ax2 - bar plot
ax2.bar(x2, y2, color='orange', edgecolor='black')
ax2.set_title('Bar Chart: Categories', fontsize=12, color='darkred')
ax2.set_xlabel('Category', fontsize=10)
ax2.set_ylabel('Value', fontsize=10)
ax2.annotate('Highest',
             xy=(x2[y2.index(max(y2))], max(y2)),
             xytext=(x2[y2.index(max(y2))], max(y2) + 0.5),
             ha='center',
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=8)

plt.tight_layout()
plt.show()
```
