```python
import seaborn as sns
import pandas as pd

# Load the 'tips' dataset
tips = sns.load_dataset("tips")

# Display the dataset
print(tips.head())
```

```python
# Create a line plot
tipgraph = sns.lineplot(
    data=tips,
    x='total_bill',
    y='tip',
    color='hotpink',
    linestyle='--',
    linewidth=3,
    marker='o',
    markerfacecolor='indigo'
)

# Customize the plot
tipgraph.set(
    title='Tips vs Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)'
)

# Show the plot
plt.show()
```

```python
# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 3))

# Create a line plot
sns.lineplot(
    data=tips,
    x='total_bill',
    y='tip',
    color='gray',
    linestyle='--',
    linewidth=3,
    marker='o',
    markerfacecolor='indigo',
    ax=ax
)

# Customize the plot
ax.set_title('Tips vs Total Bill')
ax.set_xlabel('Total Bill ($)')
ax.set_ylabel('Tip Amount ($)')

# Show the plot
plt.show()
```

```python
# Create a scatter plot
tipgraph = sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    style='time',
    hue='day',
    palette=['purple', 'hotpink', 'deepskyblue', 'yellowgreen']
)

# Customize the plot
tipgraph.set(
    title='Tips vs Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)'
)

# Show the plot
plt.show()
```

```python
daysplot = sns.relplot(
 data=tips,
 x="total_bill",
 y="tip",
 hue="sex",
 col="day",
 kind="scatter",
 col_wrap=2)
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Check if WordCloud is imported correctly
print("WordCloud imported successfully")

# Load the CSV file
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv", on_bad_lines='skip')

# Display the dataframe
print("Data loaded successfully")
print(df.head())

# Combine all quotes into a single string
text = ' '.join(df['quote'].astype(str))

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Check if the word cloud is generated
print("WordCloud generated successfully")

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

```

```python
from matplotlib_venn import venn2

# Define the sets
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])

# Create the Venn diagram
diagram = venn2([A, B],
                set_labels=('Fruits', 'Pies'),
                set_colors=('blue', 'red'),
                alpha=0.5)

# Annotate the diagram
diagram.get_label_by_id("10").set_text("\\n".join(A - B))  # Elements only in A
diagram.get_label_by_id("11").set_text("\\n".join(A & B))  # Elements in both A and B
diagram.get_label_by_id("01").set_text("\\n".join(B - A))  # Elements only in B
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

# Example data
set1 = {"A", "B", "C", "D"}
set2 = {"C", "D", "E", "F"}

# Create a Venn diagram
venn2([set1, set2], ('Set1', 'Set2'))
venn2_circles([set1, set2], linestyle='dashed')
venn2_unweighted([set1, set2], ('Set1', 'Set2'))

plt.show()
```
